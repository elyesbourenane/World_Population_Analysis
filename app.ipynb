```python
df.drop(["CCA3", "Country",	"Capital", "Continent"], axis=1).corr()
```

```python
sns.heatmap(df.drop(["CCA3", "Country",	"Capital", "Continent"], axis=1).corr(), annot=True, cmap="YlGnBu")
plt.rcParams["figure.figsize"] = (15, 10)
plt.show()
```

```python
df2 = df[['Continent', '2022 Population',
       '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population',
       '1970 Population', 'Area (km²)', 'Density (per km²)', 'Growth Rate',
       'World Population Percentage']]
df2.groupby("Continent").mean().sort_values(by="2022 Population", ascending=False)
```

```python
# visualize the continents
population_cols = ['2022 Population',
       '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population',
       '1970 Population']
population_cols.reverse()


df3 = df2.groupby("Continent")[population_cols].mean().sort_values(by="2022 Population", ascending=False)
df3 = df3.transpose()
df3.plot()
```

```python
df2.boxplot(figsize=(20, 10))
```

```python
# The evolution of the population in Algeria
algeria = df[df["Country"] == "Algeria"][population_cols].to_numpy()
plt.plot(population_cols, algeria[0])
```

```python
# The evolution of the populatio in France
france = df[df["Country"] == "France"][population_cols].to_numpy()
plt.plot(population_cols, france[0])
```
